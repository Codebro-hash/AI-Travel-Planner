```python

# Mini Project
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from google.colab import userdata

cityy = str(input("Enter the city name :"))
dayss = int(input("Enter the number of days :"))
travel_stylee = str(input("Enter thetravel style (relaxed/busy) :"))
interestss = str(input("Enter the interests :"))

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",google_api_key=userdata.get('GEMINI_API_KEY'))

travel_template = PromptTemplate(template="""generate a day-wise travel itinerary using {city} {days} {travel_style} {interests},should do the following
1) Create a Day-wise plan
Must generate the plan as:

Day 1
Day 2
â€¦ up to the number of days entered by the user
2) Include time blocks for every day
Each day must contain:

Morning
Afternoon
Evening
3) Add one budget-saving tip per day
At the end of each day, Gemini must add:

Budget Tip: (one practical money-saving suggestion)""", input_variables=["city","days","travel_style","interests"])


travel_chain = travel_template | llm | StrOutputParser()

#invoke
response = travel_chain.invoke({"city":cityy,"days":dayss,"travel_style":travel_stylee,"interests":interestss})

print(response)
```
